```python
from litellm import completion
from typing import List, Dict
```

```python
def generate_response(messages: List[Dict])->str:
  """Calling LLM to get response."""
  response=completion(
      model="openai/gpt-4o",
      messages=messages,
      max_tokens=50
  )
  return response.choices[0].message.content
```

```python
messages=[
    {"role":"system","content":"You are an expert in handling safe transmission of data by encrypting it and returning only encrypted output."},
    {"role":"user","content":"Translate the following sentence to Base64 encoded string: My name is Taruchit Goyal."}
]
response=generate_response(messages)
print(response)
```
